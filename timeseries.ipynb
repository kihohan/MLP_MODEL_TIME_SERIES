```python
import datetime
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('buzz_1614736670321.xlsx')
df['구분'] = df['구분'].astype('str')
df['구분'] = pd.to_datetime(df['구분'])
df['년'] = df['구분'].apply(lambda x:x.year)
df['달'] = df['구분'].apply(lambda x:x.month)
df['주차'] = df['구분'].apply(lambda x:(x.day - 1) // 7 +1).astype('str')
df['구분'] = df['구분'].astype('str')
```

```python
df['년주차'] = df['구분'].apply(lambda x:x.split('-')[0]) + '.' + df['구분'].apply(lambda x:x.split('-')[1]) + '.' + df['주차'] + 'W'
```

```python
data = df[['년주차','삼성 비스포크','LG 오브제컬렉션']]
```

```python
r = data.groupby('년주차')['삼성 비스포크','LG 오브제컬렉션'].sum().reset_index()
```

```python
r.to_excel('timeseries_buzz.xlsx', index = False)
```
